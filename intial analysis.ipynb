```python
l1 = []
l2 = []
l3 = []
l4 = []
l5 = []
l6 = []
l7 = []
l8 = []

def convert1(param):
    sum = 0
    for i in range(0, 12):
        sum += glob.loc[i+param].iloc[1]

    return sum/12

def convert2(param):
    sum = 0
    for i in range(0, 12):
        sum += glob.loc[i+param].iloc[2]

    return sum/12

def convert3(param):
    sum = 0
    for i in range(0, 12):
        sum += glob.loc[i+param].iloc[3]

    return sum/12

def convert4(param):
    sum = 0
    for i in range(0, 12):
        sum += glob.loc[i+param].iloc[4]

    return sum/12

def convert5(param):
    sum = 0
    for i in range(0, 12):
        sum += glob.loc[i+param].iloc[5]

    return sum/12

def convert6(param):
    sum = 0
    for i in range(0, 12):
        sum += glob.loc[i+param].iloc[6]

    return sum/12

def convert7(param):
    sum = 0
    for i in range(0, 12):
        sum += glob.loc[i+param].iloc[7]

    return sum/12

def convert8(param):
    sum = 0
    for i in range(0, 12):
        sum += glob.loc[i+param].iloc[8]

    return sum/12

for i in range(1200, 3192, 12):
    l1.append(convert1(i))
    l2.append(convert2(i))
    l3.append(convert3(i))
    l4.append(convert4(i))
    l5.append(convert5(i))
    l6.append(convert6(i))
    l7.append(convert7(i))
    l8.append(convert8(i))
```

```python
dt = 1850
date = []
for i in range(1200, 1366):
    date.append(1850 + i - 1200)
```

```python
date
```

```python
dix = {"date":date, "LandAverageTemperature":l1, "LandAverageTemperatureUncertainty":l2, "LandMaxTemperature":l3, "LandMaxTemperatureUncertainty":l4,                
       "LandMinTemperature":l5, "LandMinTemperatureUncertainty": l6, "LandAndOceanAverageTemperature":l7, "LandAndOceanAverageTemperatureUncertainty":l8}

modified = pd.DataFrame(dix)
```
